Using the Press-Schechter formalism and its modification for ellipsoidal collapse (eq. 7.45 and 7.67 in the book), integrate numerically the number densities of dark matter halos: 
differential $M \frac{dn}{dM}$ and cumulative $n(>M)=\int\frac{dn}{dM}dM$, in the range $10^{10} M_\odot < M < 10^{15} M_\odot$ at $z=0$.

The fluctuation power spectrum $P(k)$ (in units of $(h^{-1} \rm{Mpc}^3$) for the concordance cosmological model ($\Omega_{m0} = 0.3, \Omega_\Lambda = 0.7, \Omega_{b0} = 0.046, h=0.7$) is given in the second column of the table at (url). The first column in the table is comoving $k$, in $h$ Mpc$^{-1}$. The rms fluctutation within a comoving radius $R$ is 

$$ \sigma(R)^2 = \frac{\pi}{(2\pi)^3} \int_0^\infty P(k) W(kr)^2 k^2 dk,~~~~~ W(x) = 3 \frac{\sin x - x \cos x}{x^3}.$$

```python
from __future__ import division

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz

import astropy.units as u
import astropy.constants as c

pk_filename = 'pk.dat'

pk_dat = np.loadtxt(pk_filename)

k_array = pk_dat[:,0]
P_array = pk_dat[:,1]

plt.plot(k_array, P_array)
plt.loglog()
plt.xlabel("k")
plt.ylabel("P")
```

## 1.
Interpolate log $P(k)$ vs. log $k$ and integrate numerically $\sigma^2(R)$ for a grid of radii $R$ (in $h^{-1}$ Mpc) corresponding to masses $M=\frac{4\pi}{4}R^3 \bar{\rho}_{m0}$. 
$P(k)$ has arbitrary normalization in the table.
Rescale $P(k)$ such that $\sigma^2(R=8 h^{-2} \rm{Mpc}) = \sigma_8^2$, where $\sigma_8 = 0.82$.
In practice, you need to do the integral in the range $k_{min}$ to $k_{max}$. 
The values of $k$ provided in the table should be adequate for the mass range of interest.

```python
# Let's take a few steps here. We need to define a non-normalized sigma function...
# and then normalize it.

little_h = u.def_unit(['h70', 'littleh'])

def window_function(x):
    return 3 * (np.sin(x) - x*np.cos(x)) / x**3

W = window_function

def sigma_R_un_normalized(radius, normalization=1):
    
    R = u.Quantity(radius, u.Mpc/little_h)
    
    integrand = P_array * W(R.value*k_array)**2 * k_array**2 
    integral = trapz(integrand, k_array)
    
    sigma_squared = 4*np.pi / (2*np.pi)**3 * integral
    
    return normalization * (sigma_squared)**(1/2)

normalization_value = 0.82 / sigma_R_un_normalized(8)

# redefine sigma_R with the normalization
sigma_R = lambda R : sigma_R_un_normalized(R, normalization=normalization_value)
assert sigma_R(8) == 0.82
```

```python
Omega_matter = 0.3
rho_bar_m0 = 2.78e11 * Omega_matter * little_h**2 * u.Msun * u.Mpc**-3

def radius_from_mass(mass):
    
    M = u.Quantity(mass, u.Msun)
    R = (3/(4*np.pi) * M/little_h / rho_bar_m0)**(1/3)
    
    return R.to(u.Mpc/little_h)

print radius_from_mass(1e10)
    
```

```python
mass_array = np.logspace(10,15, 1000) * u.Msun

sigma_m_array = np.zeros(len(mass_array))

for i, m in enumerate(mass_array):
    
    sigma_m_array[i] = sigma_R(radius_from_mass(m))
    
plt.plot(mass_array, sigma_m_array)
plt.semilogx()

plt.xlabel(r"Mass ($M_\odot$)", fontsize=18)
plt.ylabel(r"$ \sigma(M)$", fontsize=18)
```

## 3.
Calculate $n(>M)$ by integrating the Press-Schechter expression (eq. 7.45).
Plot the resulting mass functions, cumulative $n(>M)$ and differential $M \frac{dn}{dM}$ vs. $M$, on a log-log scale *on the same graph*.
Divide or multiply by $h$, as needed, only in this final step.

Eq. 7.45:

$$ n(M,t) dM = \sqrt{\frac{2}{\pi}} \frac{\bar{\rho}}{M^2} \frac{\delta_c}{\sigma} \exp\left(- \frac{\delta_c^2}{2\sigma^2} \right) \left| \frac{d \ln \sigma}{d \ln M}\right| dM $$

which can also be written

$$ n(M,t)dM = \frac{\bar{\rho}}{M^2} f_{PS}(\nu) \left| \frac{d \ln \nu}{d \ln M}\right| dM $$
with
$$\nu = \frac{\delta_c}{\sigma(M)}~~~~~~~~~ 
f_{PS}(\nu) = \sqrt{\frac{2}{\pi}} \nu \exp\left(- \frac{\nu^2}{2} \right) $$

```python
def press_schechter_integrand(mass_array, sigma_array, delta_c=1.686):
    
    # we adjust lengths by one to accomodate the differential term, which is shortened as such.
    M = u.Quantity(mass_array[1:], u.Msun)
    sigma = sigma_array[1:]
    
    first_term = (2/np.pi)**(1/2) * rho_bar_m0/M**2 * delta_c/sigma
    second_term = np.exp(-delta_c**2/(2*sigma**2)) * np.abs(differential_of_sigma_m)
    
    nM = first_term * second_term
    
    return nM

nM = press_schechter_integrand(mass_array, sigma_m_array)
dM = np.ediff1d(mass_array)

from scipy.integrate import cumtrapz

print len(nM)
print len(dM)

# we adjust lengths by one to accomodate the nM.
cumulative_mass_function = cumtrapz(-nM[::-1], mass_array[1:][::-1])[::-1]
differential_mass_function = nM * mass_array[1:]

# we adjust lengths by two to accomodate the cumulative mass function.
plt.plot(mass_array[2:], cumulative_mass_function, '-', label="cumulative mass function")
plt.plot(mass_array[1:], differential_mass_function, '--', label="differential mass function")
plt.loglog()
plt.legend()
plt.xlabel(r"Mass ($M_\odot$)", fontsize=18)
plt.ylabel("Mass function value", fontsize=18)
```

## 5. 
Repeat the mass function calculation using the ellipsoidal collapse modification (eq. 7.67), and make a new corresponding plot of $n(>M)$ and $M \frac{dn}{dM}$.

Also add to this new graph a stellar mass function from Bell et al. (2003):

$$ \frac{dn}{dM_s} dM_s = \phi_0 \left( \frac{M_s}{M_0} \right)^{-\alpha} e^{-M_s /M_0} \frac{dM_s}{M_0} $$

with $\phi_0 = 4\times 10^{-3} \rm{Mpc}^{-3}$, $\alpha=1.1$, $M_0=7\times10^{10} M_\odot$, and its cumulative version $n(>M_s)$.
To put the halo and stellar masses on the same graph, assume $M_s = f_b M$, with $f_b = \Omega_{b0} / \Omega_{m0}$.

```python
def dn_Bell(mass_array):
    
    M_s = u.Quantity(mass_array, u.Msun)
    
    phi = 4e-3 * u.Mpc**(-3)
    alpha = 1.1
    M_0 = 7e10 * u.Msun
    
    dn = phi * (M_s/M_0)**(-alpha) * np.exp(-M_s/M_0) / M_0
    
    return dn
```

Eq. 7.67:
As above, but replace
$$ f_{PS}(\nu) => f_{EC}(\nu) = A \left(1 + \frac{1}{\tilde{\nu}^{2q}}\right) f_{PS}(\tilde{\nu})$$

where $\tilde{\nu} = 0.84 \nu$ and $q=0.3$, and $A=0.322$.

Now:
$$ n(M,t)dM = \frac{\bar{\rho}}{M^2} f_{EC}(\nu) \left| \frac{d \ln \nu}{d \ln M}\right| dM $$
with
$$\nu = \frac{\delta_c}{\sigma(M)}~~~~~~~~~ 
f_{PS}(\nu) = \sqrt{\frac{2}{\pi}} \nu \exp\left(- \frac{\nu^2}{2} \right) $$

```python
delta_c = 1.686
nu_array = delta_c/sigma_m_array
differential_of_nu = np.ediff1d(np.log(nu_array))/np.ediff1d(np.log(mass_array.value))

def f_PS(nu):
    f = np.sqrt(2/np.pi) * np.exp(-nu**2/2)
    return f

def f_EC(nu):
    nu_squiggle = 0.84*nu
    A=0.322 
    q=0.3
    f = A * (1+1/nu_squiggle**(2*q)) * f_PS(nu_squiggle)
    return f

def ellipsoidal_collapse_integrand(mass_array, nu_array):
    M = u.Quantity(mass_array[1:], u.Msun)
    nu = nu_array[1:]
    
    nM = rho_bar_m0/M**2 * f_EC(nu) * differential_of_nu
    
    return nM


nM_EC = ellipsoidal_collapse_integrand(mass_array, nu_array)

# we adjust lengths by one to accomodate the nM.
cumulative_mass_function_EC = cumtrapz(-nM_EC[::-1], mass_array[1:][::-1])[::-1]
differential_mass_function_EC = nM_EC * mass_array[1:]

```

```python
# bell stuff

Omega_baryon = 0.046
baryon_fraction = Omega_baryon/Omega_matter

stellar_mass_array = baryon_fraction * mass_array

nMs_Bell = dn_Bell(stellar_mass_array)

# we adjust lengths by one to accomodate the nM.
cumulative_Bell = cumtrapz(-nMs_Bell[::-1], stellar_mass_array[::-1])[::-1]
differential_Bell = nMs_Bell * stellar_mass_array

```

```python
# plot it all

plt.plot(mass_array[2:], cumulative_mass_function, 'b-', label="cumulative, PS")
plt.plot(mass_array[1:], differential_mass_function, 'b--', label="differential, PS")
plt.plot(mass_array[2:], cumulative_mass_function_EC, 'r-', label="cumulative, EC")
plt.plot(mass_array[1:], differential_mass_function_EC, 'r--', label="differential, EC")

plt.loglog()
plt.plot(mass_array[1:], cumulative_Bell, 'g-', label="cumulative Bell", scaley=False, scalex=False)
plt.plot(mass_array, differential_Bell, 'g--', label="differential Bell", scaley=False, scalex=False)
leg = plt.legend(loc='lower left')
leg.get_frame().set_alpha(0.3)
plt.xlabel(r"Mass ($M_\odot$)", fontsize=18)
plt.ylabel("Mass function value", fontsize=18)
```

## 6.
Comment on the similarities and differences between the mass functions of halos and galaxies.

Similarities: they are all monotonically decreasing.
Differences: the "knee" of the stellar mass functions is much more pronounced. The slope for stars, post-knee is very steep, and rather shallow pre-knee.

The "baryonic mass function" should in principle follow the halo mass function, since baryons are just a constant fraction of matter.
However, the stellar/galaxy mass functions do not follow this, so something must be operating on the baryons before they become stars.
I would postulate that the processes that separate the stellar mass function from the halo mass function are:
* at low mass, supernova feedback prevents gas to efficiently form stars
* at high mass, AGN feedback prevents gas to efficiently form stars.

## 7.
Using the abundance matching method, $n(>M) = n(>M_s)$, calculate the stellar mass of a galaxy associated with halo of mass $M$ (that is, not assuming $M_s = f_b M$). Make a plot of the ratio $M_s/M$ vs. $M$.

```python
def match_abundance(mass_array, stellar_mass_function, halo_mass_function):
    
    stellar_mass_of_halo = np.zeros(len(mass_array))
    
    for i, m in enumerate(mass_array[1:]):
        
        indices_with_overlap = mass_array[stellar_mass_function.value >= halo_mass_function[i].value]
        
        if len(indices_with_overlap) > 0:
            stellar_mass_of_halo[i] = np.max(indices_with_overlap).value
    
    return stellar_mass_of_halo

stellar_mass_of_halos = match_abundance(mass_array, nMs_Bell, nM_EC)
print len(stellar_mass_of_halos)

plt.plot(mass_array, stellar_mass_of_halos/mass_array)
plt.semilogx()
plt.xlabel("Halo mass ($M_\odot$)", fontsize=18)
plt.ylabel("Stellar mass fraction of halo $M_s/M$", fontsize=14)
```
